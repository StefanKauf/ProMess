```python
"""
Date: 12.01.2023
Author: Kaufmann Stefan

Labor Prozessmesstechnik
"""
# Load Robotermodell
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# update latex preamble
plt.rcParams.update({
    "font.family": "serif",
    "text.usetex": True,
    "pgf.rcfonts": False,
    "pgf.texsystem": 'pdflatex', # default is xetex
    "pgf.preamble": [
         r"\usepackage[T1]{fontenc}",
         r"\usepackage{mathpazo}"
         ]
})
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')



# importieren der Messdaten 
Z = pd.read_excel("Daten/Impedanzgang.xlsx")             # mit R = 51
Z_u = pd.read_excel("Daten/Impedanz_unten.xlsx")         # mit R = 51
Z_R2 = pd.read_excel("Daten/Impedanzgang_R2200.xlsx")    # mit Wiederstand R = 2k2



plt.figure()
plt.plot(Z.values[:,0], Z.values[:,1], 'b', label='\(| Z |\) ')
plt.legend(loc='best')
plt.xlabel('Frequnez [Hz]')
plt.ylabel('\(| Z |\) [$\Omega$]' )
plt.grid()
plt.yscale('log')
plt.xscale('log')
plt.title("Impedanzgang")
plt.grid(linewidth = 0.5)
plt.legend()
plt.savefig("Impedanzgang.svg")



plt.figure()
plt.plot(Z_u.values[:,0], Z_u.values[:,1], 'b', label='\(| Z |\),  R= 51 $\Omega$  ')
plt.plot(Z_R2.values[:,0], Z_R2.values[:,1], 'g', label='\(| Z |\) , R= 2k2 $\Omega$')
plt.legend(loc='best')
plt.xlabel('Frequnez [Hz]')
plt.ylabel('\(| Z |\) [$\Omega$]' )
plt.grid()
plt.yscale('log')
plt.xscale('log')
plt.title("Impedanzgang unterhalb der ersten Resonanz")
plt.grid(linewidth = 0.5)
plt.legend()
plt.savefig("Impedanzgang_unten.svg")

```
