```python
import numpy as np
import matplotlib.pyplot as plt

Q   = 80
k   = 0.7
A   = (25e-03)**2*np.pi
t   = 2e-03
q   = 7.7e-03
e33 = 1800e-12
#f   =   np.array([[1e03, 1e04, 1e05]])
f = np.linspace(1000, 100*1000, 1000)  # Sample data.
w   = 2*np.pi*f

L  =  q*A*t/k**2 *1e06

C0  =   e33*A/t
C = C0*k**2
R = 17e-3/(w*C)


ReZ = R
ImZ = L*w-1/(C*w)
ReN = 1-w*C0*(L*w-1/(C*w))
ImN = w*C0*R
betrag = (ReZ**2+ImZ**2)**0.5/(ReN**2+ImN**2)**0.5



plt.plot(f, betrag, label='Impedanzgang') 
plt.yscale("log")
plt.xscale("log")
plt.xlabel('Frequnez')
plt.ylabel('|Z|')
plt.title("Impedanzgang")
plt.grid()
plt.legend()
```
